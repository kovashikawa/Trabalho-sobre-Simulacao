Para o primeiro tipo de cliente para um fração **f** = 50%:

Temos que $a = -0.55$, então: $$ f_X(x) = (0.45)x^{-0.55} ,\ \forall x \in (0,1) ,$$

Então, $$Y = 2 + 2X.$$

Logo, temos que: $f_Y(y)= \frac{f_X(y)}{g'(y)} \iff f_X(y)=(.45)\big(1+ \frac {y}{2}\big)^{-.55} ;\ g'(y)= 2$ 

$$\Longrightarrow f_Y(y)= (0.225)\Bigg(1+ \frac {y}{2}\Bigg),\ \forall y \in (2,4)$$

***

Para o segundo tipo de cliente:

Temos que $\lambda = 5.5$, tal que $\mu = \frac {1}{\lambda} \Longrightarrow \mu = \frac{1}{5.5} $,

então: 

$$ f_Z(z) = \frac{e^{ \frac{\Large-z\Large}{5.5}}}{5.5},\ \forall z \in [0,\infty )$$

como __f__ $= 0.5$, teremos uma $f.d.p$ de $W \sim Bernoulli(0.5)$ que distribuirá a quantidade de clientes _demorados (com $Y \sim Poisson(10)$) ou rápidos (com $Z \sim Expo(\frac {1}{5.5})$)_

Outrossim, $$ W \sim Binomial(1, 0.5) $$

````python
# Tempo de Serviço

tmp_serv = []

for cliente in range(10000):
    
    if simulacao['Cliente Rápido?'][cliente] == 1:
        
        tmp_serv.append(rapid[cliente])
        
    else:
        
        tmp_serv.append(slow[cliente])

simulacao['Tempo de Serviço'] = tmp_serv

simulacao.tail()
````

````python
# Início do Atendimento

inicio = [instante[0]]

for cliente in range(1,10000):
    
    i = inicio[cliente-1] + tmp_serv[cliente-1]
    
    if i <= instante[cliente]:
        
        inicio.append(instante[cliente])
    else:
        
        inicio.append(i)
    

simulacao['Início do Atendimento'] = inicio

simulacao.head()
````

````python
# Fim do Atendimento

fim = []

for cliente in range(10000):
    
    i = inicio[cliente] + tmp_serv[cliente]
    fim.append(i)

simulacao['Fim do Atendimento'] = fim

simulacao.head()
````

````python
# Tempo de Espera

espera = []

for cliente in range(10000):
    
    if inicio[cliente] <= instante[cliente]:
        
        espera.append(0)
        
    else:
        
        espera.append(inicio[cliente] - instante[cliente])

simulacao['Tempo de Espera'] = espera

# Tempo Ocioso

ocioso = [inicio[0]]

for cliente in range(1,10000):
    if inicio[cliente] <= fim[cliente - 1]:
        ocioso.append(0)
    else:
        ocioso.append(inicio[cliente] - fim[cliente-1])

#simulacao['Tempo de Espera'] = espera

#simulacao.head()

simulacao['Tempo Ocioso'] = ocioso

simulacao.head()
````

### Questão 2 

    Quantas figurinhas devem ser compradas até completar um álbum com N figurinhas, supondo que não haja trocas de figurinhas? Simule 100 repetições deste processo, supondo que as figurinhas sejam compradas individualmente e que não haja figurinhas difíceis.
    
    a)   Calcule a média e o desvio padrão obtido nas 100 realizações e transcreva-os para a planilha de resultados (somente os valores).  Obtenha também o valor exato da média e do desvio-padrão e compare com o resultado da simulação. 
    
    b)   Usando a simulação, estime a probabilidade de que o número de figurinhas compradas seja maior que M.
   
__Solucao:__

temos que $N = 635$ e $M = 4500$, usando uma tabela (*pandas.Dataframe*)

#### Letra a)

````python
# criando 100 simulacoes:

columns = ['Figurinha']

for i in range(1,101):
    
    s = 'Sim {}'.format(i)
    columns.append(s)
    
columns.append('Valores')
columns.append('Teóricos')
# criando o DataFrame:

Figurinhas = pd.DataFrame(columns=columns)

fig = range(1,636)

Figurinhas['Figurinha'] = fig

Figurinhas.tail()
````

````python
import math

for i in range(1,101): # as colunas de Sim 1 a 100
    
    sim_h = [1]
    
    for figurinha in range(1,635):

        f = math.ceil(np.log(1-np.random.rand())/np.log(fig[figurinha - 1]/635))

        sim_h.append(f)

    
    Figurinhas['Sim {}'.format(i)] = sim_h

Figurinhas
````

````python
Sims = []

for i in range(1,101):
    
    s = 'Sim {}'.format(i)
    Sims.append(s)

# pegando as medias e desvios padrao das simulacoes
    
media_std = Figurinhas[Sims].describe()

media_std = media_std[1:3]

media_std
````

````python
valores = []

teoricos = []

for fig in range(1,636):
    
    val = 635/(636-fig)
    valores.append(val)
    
    teo = 635*(fig-1)/(636-fig)**2
    teoricos.append(teo)
    
Figurinhas['Valores'] = valores
Figurinhas['Teóricos'] = teoricos
````

````python
Figurinhas
````

````python
# calculando as medias

media_hat = Figurinhas[Sims].sum(axis=0).mean()

media = Figurinhas['Valores'].sum(axis=0)

print('A média de figurinhas na simulação é', media_hat,
     '\nA média de figurinhas na teoria é', media)
````

````python
# calculando os STDs

std_hat = np.std(Figurinhas[Sims].sum())

std = math.sqrt(Figurinhas['Teóricos'].sum())

print('O desvio-padrao de figurinhas na simulação é', std_hat,
     '\nO desvio-padrao de figurinhas na teoria é', std)
````

#### Letra b)

````python
favoravel = 0

for figs in Figurinhas[Sims].sum(axis=0):

    if figs > 4500:
    
        favoravel = favoravel +1
````

````python
print('Então,\n\t P(No. fig>M) =',favoravel/100)
````
